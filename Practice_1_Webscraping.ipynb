## Scraping for Only one Page

```python
doc = 'https://www.homecredit.co.id/Spesifikasi-dan-Harga-Xiaomi-Pocophone-F2'

#define html parser from beautiful soup
response = requests.get(doc, timeout = 10)
soup = bs(response.content,"html.parser")

#get the headline element
title = soup.find('title')
title = title.text.strip()

#get the content
content = [element.text for element in soup.find_all('div',{'itemprop':'articleBody'})]
print('title:', title)
print('content:',content)
```

```python
doc = 'https://blog.kredivo.com/1818-2/'

#define html parser from beautiful soup
response = requests.get(doc, timeout = 10)
soup = bs(response.content,"html.parser")

#get the headline element
title = soup.find('title')
title = title.text.strip()

#get the content
content = [element.text for element in soup.find_all('div',{'class':'entry-content'})]
print('title:', title)
print('content:',content)
```

```python
doc = 'https://blog.bukalapak.com/inovasi/product-guarantee-cara-belanja-online-bebas-was-was-di-bukalapak-104059'

#define html parser from beautiful soup
response = requests.get(doc, timeout = 10)
soup = bs(response.content,"html.parser")

#get the headline element
title = soup.find('title')
title = title.text.strip()

#get the content
content = [element.text for element in soup.find_all('div',{'class':'article-body'})]
print('title:', title)
print('content:',content)
```

```python
html_docs = ['https://www.homecredit.co.id/Spesifikasi-dan-Harga-Xiaomi-Pocophone-F2',
             'https://www.homecredit.co.id/5-Fakta-Menarik-Sadio-Mane',
             'https://www.homecredit.co.id/Smartphone-Gaming-Murah-Terbaik',
             'https://www.homecredit.co.id/5-Cara-Mengatur-Keuangan-Rumah-Tangga']
#extract documents and store in one array
docs = []
for index, document in enumerate (html_docs):
    
    #prepare a dictionary for each page
    doc1 = {}
    #define html parser from beautiful soup
    response = requests.get(document, timeout = 10)
    soup = bs(response.content,"html.parser")
    
    #clean the text
    title = soup.find ('title')
    doc1['id'] = index + 1
    doc1['title'] = title.text.strip() #clean the text

    #store all the rows within div element
    content = [ element.text for element in soup.find_all('div',{'itemprop':'articleBody'})]
#remove unwanted lines
    doc1['content'] = ''.join(element for element in content)

    docs.append(doc1)
    print(docs)
```

```python
#create a dataframe from above list
data = pd.DataFrame(docs, columns = ['id','title','content'])
data
```

lets try to get data for Apple's stock with 30 minutes interval

```python
API_URL = 'https://www.alphavantage.co/query'
def extract_data(symbol, interval, API_URL, api_key):
    data = {'function':'TIME_SERIES_INTRADAY',
            'symbol':symbol,
            'interval':interval,
            'datatype':'json',
            'apikey':api_key}
    
    response = requests.get(API_URL, data)
    data = response.json()
    data = data['Time Series (%s)' % (interval)]
    
    df = pd.DataFrame.from_dict(data, orient='index')
    
    return df
```

```python
#lets try to get data for apple stock with 30 minutes interval

#extract data to dataframe
aapl = extract_data('AAPL', '30min', API_URL, api_key)
aapl
```
